# <p style="background-color:#EE82EE;font-family:newtimeroman;color:black;font-size:200%;text-align:center;border-radius:10px 10px;"> "  Data Visualization  " </p>

```python
plt.figure(figsize=(20,5))
plt.title('number of medals over the Year')
df.Year.value_counts().sort_index().plot()
```

```python
plt.figure(figsize=(20,5))
plt.title('number of medals over the Games')
df.Games.value_counts().sort_index().plot()
```

```python
plt.figure(figsize=(20,5))
plt.title('number of medals over the Age')
df.Age.value_counts().sort_index().plot()
```

```python
plt.figure(figsize=(20,5))
plt.title('number of medals over the Height')
df.Height.value_counts().sort_index().plot()
```

```python
plt.figure(figsize=(20,5))
plt.title('number of medals over the Weight')
df.Weight.value_counts().sort_index().plot()
```

```python
itemNames = df['Name'].value_counts().index[:8]
itemValues = df['Name'].value_counts().values[:8]
plt.figure(figsize=(20,9))
plt.ylabel('count', fontsize='medium')
plt.xlabel('Name', fontsize='medium')
plt.title('Top 10 Name')
plt.bar(itemNames,itemValues, width = 0.7,color='purple',linewidth=0.4)
for i in range(len(itemNames)):
    plt.text(i,itemValues[i],itemValues[i],ha='center',va='bottom')
plt.show()
```

```python
itemNames = df['Team'].value_counts().index[:8]
itemValues = df['Team'].value_counts().values[:8]
plt.figure(figsize=(20,9))
plt.ylabel('count', fontsize='medium')
plt.xlabel('Team', fontsize='medium')
plt.title('Top 10 Team')
plt.bar(itemNames,itemValues, width = 0.7,color='purple',linewidth=0.4)
for i in range(len(itemNames)):
    plt.text(i,itemValues[i],itemValues[i],ha='center',va='bottom')
plt.show()
```

```python
itemNames = df['Games'].value_counts().index[:8]
itemValues = df['Games'].value_counts().values[:8]
plt.figure(figsize=(20,9))
plt.ylabel('count', fontsize='medium')
plt.xlabel('Games', fontsize='medium')
plt.title('Top 10 Games')
plt.bar(itemNames,itemValues, width = 0.7,color='purple',linewidth=0.4)
for i in range(len(itemNames)):
    plt.text(i,itemValues[i],itemValues[i],ha='center',va='bottom')
plt.show()
```

```python
itemNames = df['Sport'].value_counts().index[:8]
itemValues = df['Sport'].value_counts().values[:8]
plt.figure(figsize=(20,9))
plt.ylabel('count', fontsize='medium')
plt.xlabel('Sport', fontsize='medium')
plt.title('Top 10 Sport')
plt.bar(itemNames,itemValues, width = 0.7,color='DeepPink',linewidth=0.4)
for i in range(len(itemNames)):
    plt.text(i,itemValues[i],itemValues[i],ha='center',va='bottom')
plt.show()
```

```python
labels = df.Sex.value_counts().index
colors = ['royalblue','orangered']
Sex = df.Sex.value_counts().values
plt.figure(figsize = (10,10))
plt.pie(Sex, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('proportion of Sex',color = 'black',fontsize = 30)
```

```python
labels = df.Medal.value_counts().index
colors = ['royalblue','orangered']
Medal = df.Medal.value_counts().values
plt.figure(figsize = (10,10))
plt.pie(Medal, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('proportion of Medal',color = 'black',fontsize = 30)
```
